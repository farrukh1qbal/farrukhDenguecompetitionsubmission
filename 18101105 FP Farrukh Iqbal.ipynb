```python
import pandas as pd
import numpy as np
training_data = pd.read_csv('D:\FYP Tools\dengue_features_train.csv')
print('Training Data',training_data.head())
print(training_data.shape)
testing_data = pd.read_csv('D:\FYP Tools\dengue_features_test.csv')
print('Testing Data',testing_data.head())
print(testing_data.shape)
```

```python
print(training_data.dtypes)
print(testing_data.dtypes)
```

```python
print(training_data.describe(include='all'))
print(testing_data.describe(include='all'))
```

```python
print(training_data.isna().sum())
print(testing_data.isna().sum())
```

```python
train=training_data.fillna(training_data.mean())
print(train.isnull().sum(axis=0))
test=testing_data.fillna(testing_data.mean())
print(test.isnull().sum(axis=0))
```

```python
train['week_start_date'] = train['week_start_date'].astype('category') 
train['week_start_date'] = train['week_start_date'].cat.codes
train['city'] = train['city'].astype('category') 
train['city'] = train['city'].cat.codes
test['city'] = test['city'].astype('category') 
test['city'] = test['city'].cat.codes
test['week_start_date'] = test['week_start_date'].astype('category') 
test['week_start_date'] = test['week_start_date'].cat.codes
print(train.shape)
print(test.shape)
```

```python
from sklearn.model_selection import train_test_split
x_training = train.drop(['total_cases'], 1)
y_training = train['total_cases']
x_train, x_test, y_train, y_test = train_test_split(x_training, y_training, test_size=0.2855, random_state=1)
print(x_test.shape)
print(test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
regression = LinearRegression()
regression.fit(x_train, y_train)
y_predicted = regression.predict(x_test)
mae = mean_absolute_error(y_test, y_predicted)
print(mae)
```

```python
y_predicted = regression.predict(test)
print(y_predicted)
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)
RFR.fit(x_train, y_train)
y_predict = RFR.predict(x_test)
mae = mean_absolute_error(y_test, y_predict)
print(mae)
```

```python
y_pred = RFR.predict(test)
y_prediction=y_pred.astype(int)
print(y_prediction)
```
